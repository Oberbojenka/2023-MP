```python
from datetime import date
import itertools

# Исходные данные
fio = ('Иванов Иван Иванович', 3, 4, 2003)
school_marks = {'Математика': 5, 'Физика': 4, 'Химия': 4, 'Биология': 5, 'Русский язык': 5, 'Литература': 4, 'История': 4, 'Обществознание': 4, 'География': 5, 'Английский язык': 5, 'Французский язык': 4,'Информатика': 5}
kinsmans = ['Иван 2003', 'Ольга 1973', 'Николай 1973', 'Григорий 1998', 'Владимир 1972', 'Анастасия 2002', 'Эльвира 2000', 'Варвара 2011', 'Дарья 2011', 'Александр 1977','Татьяна 1968', 'Наталья 1977']
pet_name = 'Анатолий'

# Средняя оценка в аттестате
average = sum(school_marks.values()) / len(school_marks)
print(f'Средняя оценка: {average:.2f}')

# Уникальные имена среди родственников
unique_names = set(name.split()[0] for name in kinsmans) 
print(f'Уникальные имена: {", ".join(unique_names)}')

# Общая длина всех названий предметов
total_length = sum(len(subject) for subject in school_marks)
print(f'Общая длина названий предметов: {total_length}')

# Уникальные буквы в названиях предметов
unique_letters = set(char for subject in school_marks for char in subject)
print(f'Уникальные буквы в предметах: {", ".join(unique_letters)}')

# Имя домашней пушистой кивы в бинарном виде
binary_name = ''.join(format(ord(char), '08b') for char in pet_name)
print(f'Имя домашней пушистой кивы в бинарном виде: {binary_name}')

# Отсортированный по алфавиту и в обратном порядке список родственников
kinsmans1 = kinsmans.copy()
kinsmans1=list(name.split()[0] for name in kinsmans)
sorted_kinsmans1 = sorted(kinsmans1, reverse=True)
print(f'Отсортированный по алфавиту и в обратном порядке список родственников: {",".join(sorted_kinsmans1)} '  )

# Количество дней от даты рождения до текущей даты
birth_date = date(fio[3], fio[2], fio[1])
today = date.today()
days_since_birth = (today - birth_date).days
print(f'Количество дней от даты рождения до текущей даты: {days_since_birth}')

# Замена имени на имя ацтекского правителя
aztec_rulers = ['Acamapichtli', 'Huitzilihuitl', 'Chimalpopoca', 'Itzcoatl', 'Moctezuma I', 'Axayacatl', 'Tizoc', 'Ahuitzotl', 'Moctezuma II', 'Cuitlahuac', 'Cuauhtemoc', 'Canek']
number = (fio[1] + fio[2]**2 + fio[3]) % 21 + 1
# FIFO очередь вводимых предметов
queue = []
while True:
    index = input('Введите индекс предмета или "стоп" для выхода:')
    if index == 'стоп':
        break
    elif a:=int(index)>10:
        print('Некорректный индекс!')
        continue     
    sorted_kinsmans1[int(index)] = aztec_rulers[number]
    subject = list(school_marks.keys())[int(index)]
    queue.append(subject)
                                    
print(f'Очередь предметов: {", ".join(queue)}')
print(f'Очередь предметов: {", ".join(sorted_kinsmans1)}')


# Связный список родственников
sorted_kinsmans = sorted(kinsmans, key=lambda x: int(x.split()[1]))
kinsman_dict = {}
for i, kinsman in enumerate(sorted_kinsmans):
    if i < len(sorted_kinsmans) - 1:
        next_index = i + 1
    else:
        next_index = None
    kinsman_dict[kinsman.split()[0]] = next_index
print(kinsman_dict)

# Функция-генератор
family_names = [kinsman.split()[0] for kinsman in kinsmans]
number = len(fio[0]) * len(family_names) % 4
def aliquot_sequence(start):
    i = start
    while True:
        if i == start:
            i = i
            start += 1
        else:
            summ= []
            for n in range(1, i + 1): 
                if i % n == 0:
                    summ.append(i // n)
            i = sum(summ) - i
        yield i
if number == 0:
    seq = aliquot_sequence(24)
    print(list(itertools.islice(seq, 10)))
```
